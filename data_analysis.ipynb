```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from src import load_train_data
```

# Analyze the data

```python
data, labels = load_train_data()

# Separate data by label
indices_by_digit = [np.where(labels == i)[0] for i in range(10)]

# Plot some examples of each digit in one figure, row by row
num_examples = 10
fig, axs = plt.subplots(10, num_examples, figsize=(num_examples * 2, 20))  # Adjust figsize as needed

for i in range(10):
    for j in range(num_examples):
        axs[i, j].imshow(data[indices_by_digit[i][j], 0], cmap='gray')
        axs[i, j].axis('off')
        if j == 0:
            axs[i, j].set_ylabel(f'Digit {i}', size='large')

plt.tight_layout()
plt.show()
```

```python
# Plot number of digits per class
plt.bar(range(10), [len(indices_by_digit[i]) for i in range(10)])
plt.xlabel('Digit')
plt.ylabel('Number of examples')
plt.show()
```

```python
# Show average number of non-zero pixels per class
plt.bar(range(10), [torch.sum(data[indices_by_digit[i], 0] > 0) / (28 * 28 * len(indices_by_digit[0])) for i in range(10)])
plt.xlabel('Digit')
plt.ylabel('Average number of non-zero pixels')
plt.show()
```
